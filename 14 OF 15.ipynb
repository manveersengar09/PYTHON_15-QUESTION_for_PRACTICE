7. Take a string and check whether it is a palindrome or not.
   

```python
str1 = input("Enter a string: ")
str2 = str1[::-1]
if str1 == str2:
    print(f"{str1} is palindrome ")
else:
    print(f"{str1} is not a palindrome")
```

6. Take a string and count how many vowels are present in it. 

```python
text = input("Enter a string: ")
count = 0

for ch in text:
    if ch.lower() in "aeiou":
        count += 1

print("Number of vowels:", count)
```

10. Take 5 numbers in an array and find the maximum number. 

```python
numbers = []

for i in range(5):
    num = int(input(f"Enter number {i+1}: "))
    numbers.append(num)

print("All elements are:", numbers)
maximum = max(numbers)
print("maximum number is:", maximum)
```

11. Take 5 numbers in an array and count how many numbers are even. 

```python
numbers = []

for i in range(5):
    num = int(input(f"Enter number {i+1}: "))
    numbers.append(num)

print("All elements are:", numbers)

count = 0
for num in numbers:
    if num % 2 == 0:
        count += 1

print("Even numbers count:", count)
```

13. Create a class Student with variables name and age. Initialize them using a constructor and print 
the values. 

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)


name = input("Enter name: ")
age = int(input("Enter age: "))

s1 = Student(name, age)
s1.display()
```
