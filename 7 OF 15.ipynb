1. Take a number from the user and check whether it is even or odd. 

```python
num = int(input("enter the number :- "))
if num % 2 == 0:
    print("number is even")
    
else:
    print("number is odd")
```

2. Take three numbers from the user and print the largest number.

```python
num1 = int(input("enter first number :- "))
num2 = int(input("enter second number :- "))
num3 = int(input("enter third number :- "))
if (num1 > num2) and (num1 > num3):
    print(f"{num1} is largest number ")
elif  (num2 > num1) and (num2 > num3):
    print(f"{num2} is largest number ")
elif (num3 > num1) and (num3 > num2):
    print(f" {num3} number is largest number ")
else:
    (print("invalid form "))
    
```

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))

largest = max(num1, num2, num3)
print("Largest number is:", largest)
```

3. Take a number and print its multiplication table till 10. 

```python
num = int(input("enter the number :- "))

for i in range(1, 11):
    print(f"{num} x {i} = {num * i}")
```

7. Take a string and check whether it is a palindrome or not.
   

```python
str1 = input("Enter a string: ")
str2 = str1[::-1]
if str1 == str2:
    print(f"{str1} is palindrome ")
else:
    print("not a palindro")
```
